```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.linspace(0,1000,1000)
y = 3*X + 5
y2 = X**2
plt.plot(X,y)
plt.plot(X,y2)
```

```python
plt.plot(X,np.log(y))
plt.plot(X,np.log(y2))
```

```python
y3 = X**3
plt.plot(np.log(X),np.log(y))
plt.plot(np.log(X),np.log(y2),c="blue")
plt.plot(np.log(X),np.log(y3),c="red")
```
